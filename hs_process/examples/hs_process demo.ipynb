## 1. Spatially crop hyperspectral datacube
### First, figure out how to crop the datacube
The Wells study has many plots (usually ~25) in each hyperspectral datacube (captured via drone at 40 m altitude). Therefore, we have to crop each datacube many times to obtain datacubes that only contain pixels representing each individual plot. To do so, we have to first record some information about the datacube and plots within that datacube.

#### Record file location
1. In a spreadsheet, record the **directory** ("directory" column) and **filename** ("fname" column) of the datacube to be processed.
2. Using software that is capable of rendering imagery, open the datacube so you are able to identify the pixel coordinates within an image (i.e., row number and column number of each pixel).
3. Locate the northwest-most plot you'd like to crop from the datacube, and record the **plot ID** ("plot_id" column), the **upper left pixel coordinate in the easting direction** (i.e, column number; "pix_e_ul" column), and the **upper left pixel coordinate in teh northing direction** (i.e, row number; "pix_n_ul" column).
4. This information is the minimum needed (in addition to a file containing the polygon features representing the plots) to be able to crop many individual plots from a full datacube. The following optional parameters may also be recorded in the spreadsheet and will be considered during the batch processing (see hs_process documentation for detailed explanation):
 - Size of the plot as you'd like it cropped in both directions (either pixel units or map units, e.g., meters); "crop_e_pix" and "crop_e_m" refer to the cropped size in the easting direction in pixel units and map units, respectively, and "crop_n_pix" and "crop_n_m" refer to the cropped size in the northing direction in pixel units and map units, respectively.
 - Size of the plot buffer (applied *after* the cropped size); "buf_e_pix" and "buf_e_m" refer to the buffer size in the easting direction in pixel units and map units, respectively, and "buf_n_pix" and "buf_n_m" refer to the cropped size in the northing direction in pixel units and map units, respectively. A positive buffer shrinks the cropped size, and a negative buffer expands the cropped size.
 - Number of plots to crop before terminating the process ("n_plots" column). This parameter can be passed to count the number of plots from the reference plot (i.e., "plot_id" column) from West to East and North to South. When the number of plots is reached, no more plots are cropped even if there are more within the extent of the datacube.

#### Load the geodata
Before actually cropping the datacube, we have to load the file containing the polygon features representing the plots. `Geopandas` will be used to read a .shp file containing the plot ID and polygon geometry (as well as other treatment information) for each plot in the experiment:

```python
import geopandas as gpd

fname_shp = r'G:\BBE\AGROBOT\Shared Work\Data\PikaImagery4_Reflectance\hs_process_demo\plot_bounds\plot_bounds.shp'
gdf = gpd.read_file(fname_shp)
gdf
```

#### Use hs_process to batch crop the datacube
There are three methods for spatially cropping a datacube, and we will use `method = many_gdf` (see the documentation for a complete description of each method). `method = many_gdf` basically overlays the datacube over the geodata to determine which plots are present in the datacube, then individually crops each plot from West to East and North to South starting with the reference plot "plot_id" column above. There are also options for modifying the output foler name, a name to append to the filename of the cropped datacube, and whether to output a geotiff of the cropped image (see the documentation for a complete list of optional parameters). The `batch.spatial_crop()` function reads the spreadsheet and conducts the cropping based on the passed parameters:

```python
from hs_process import batch

hsbatch = batch()
method = 'many_gdf'  # options: 'single', 'many_grid', or 'many_gdf'
hsbatch.spatial_crop(fname_sheet,
                     folder_name='spatial_crop_{0}'.format(method),
                     name_append='spatial-crop', geotiff=True, method=method,
                     gdf=gdf, out_force=True)
```

#### Display the cropped geotiffs
Note that the loaded geotiffs appear pretty dark. This is because of the limited rendering functionality of `rasterio`. The pixel values of the both the datacubes and the geotiffs are in units of reflectance (range from 0.0 to 1.0). 

```python
import matplotlib.pyplot as plt
import rasterio as rast
from rasterio.plot import show

base_dir = r'G:\BBE\AGROBOT\Shared Work\Data\PikaImagery4_Reflectance\hs_process_demo\spatial_crop_many_gdf'
tif_list = hsbatch._recurs_dir(base_dir, search_ext='.tif', level=0)

gdf_filter = gdf.copy()
# gdf_filter = hsbatch.my_spatial_mod._overlay_gdf(hsbatch.io.spyfile, gdf)

fig, ax = plt.subplots(figsize=(15, 15))
gdf_filter.plot(ax=ax, alpha=0.9, column='Crop', facecolor='none', edgecolor='blue', linewidth=3)

left = 1e9
bottom = 1e9
right = -1e9
top = -1e9
for fname_tif in tif_list:
    tif = rast.open(fname_tif)
    show(tif.read(), ax=ax, transform=tif.transform, adjust='linear')
    if tif.bounds[0] < left:
        left = tif.bounds[0]
    if tif.bounds[1] < bottom:
        bottom = tif.bounds[1]
    if tif.bounds[2] > right:
        right = tif.bounds[2]
    if tif.bounds[3] > top:
        top = tif.bounds[3]

# bbox = gdf_filter.total_bounds
plt.xlim((left-10, right+10))
plt.ylim((bottom-10, top+10))
```
