# Hyperspectral Image Processing Demo
## A demo of how to batch process hyperspectral imagery using `hs_process`

Note: Input datacubes are already in units of percent reflectance as pre-processing steps have already been completed.

## 1. Spatially crop hyperspectral datacube
### Figure out how to crop the datacube
The Wells study has many plots (usually ~25) in each hyperspectral datacube (captured via drone at 40 m altitude). Therefore, we have to crop each datacube many times to obtain datacubes that only contain pixels representing each individual plot. To do so, we have to first record some information about the datacube and plots within that datacube.

#### Record file location
1. In a spreadsheet, record the **directory** ("directory" column) and **filename** ("fname" column) of the datacube to be processed.
2. Using software that is capable of rendering imagery, open the datacube so you are able to identify the pixel coordinates within an image (i.e., row number and column number of each pixel).
3. Locate the northwest-most plot you'd like to crop from the datacube, and record the **plot ID** ("plot_id" column), the **upper left pixel coordinate in the easting direction** (i.e, column number; "pix_e_ul" column), and the **upper left pixel coordinate in the northing direction** (i.e, row number; "pix_n_ul" column).
4. This information is the minimum needed (in addition to a file containing the polygon features representing the plots) to be able to crop many individual plots from a full datacube. The following optional parameters may also be recorded in the spreadsheet and will be considered during the batch processing (see `hs_process` documentation for detailed explanation):
 - Size of the plot as you'd like it cropped in both directions (either pixel units or map units, e.g., meters); "crop_e_pix" and "crop_e_m" refer to the cropped size in the easting direction in pixel units and map units, respectively, and "crop_n_pix" and "crop_n_m" refer to the cropped size in the northing direction in pixel units and map units, respectively.
 - Size of the plot buffer (applied *after* the cropped size); "buf_e_pix" and "buf_e_m" refer to the buffer size in the easting direction in pixel units and map units, respectively, and "buf_n_pix" and "buf_n_m" refer to the cropped size in the northing direction in pixel units and map units, respectively. A positive buffer shrinks the cropped size, and a negative buffer expands the cropped size.
 - Number of plots to crop before terminating the process ("n_plots" column). This parameter can be passed to count the number of plots from the reference plot (i.e., "plot_id" column) from West to East and North to South. When the number of plots is reached, no more plots are cropped even if there are more within the extent of the datacube.

We've recorded this information for a single datacube and saved it to a .csv file:

```python
import pandas as pd

fname_sheet = r'G:\BBE\AGROBOT\Shared Work\Data\PikaImagery4_Reflectance\hs_process_demo\wells_image_cropping.csv'
df_crop = pd.read_csv(fname_sheet)
df_crop
```

### Load the geodata
Before actually cropping the datacube, we have to load the file containing the polygon features representing the plots. `Geopandas` will be used to read a .shp file containing the plot ID and polygon geometry (as well as other treatment information) for each plot in the experiment:

```python
import geopandas as gpd

fname_shp = r'G:\BBE\AGROBOT\Shared Work\Data\PikaImagery4_Reflectance\hs_process_demo\plot_bounds\plot_bounds.shp'
gdf = gpd.read_file(fname_shp)
gdf
```

### Use `hs_process` to batch crop the datacube
There are three methods for spatially cropping a datacube, and we will use `method = many_gdf` (see the documentation for a complete description of each method). `method = many_gdf` basically overlays the datacube over the geodata to determine which plots are present in the datacube, then individually crops each plot from West to East and North to South starting with the reference plot "plot_id" column above. There are also options for modifying the output foler name, a name to append to the filename of the cropped datacube, and whether to output a geotiff of the cropped image (see the documentation for a complete list of optional parameters). The `batch.spatial_crop()` function reads the spreadsheet and conducts the cropping based on the passed parameters:

```python
from hs_process import batch

hsbatch = batch()
method = 'many_gdf'  # options: 'single', 'many_grid', or 'many_gdf'
hsbatch.spatial_crop(fname_sheet,
                     folder_name='spatial_crop_{0}'.format(method),
                     name_append='spatial-crop', geotiff=True, method=method,
                     gdf=gdf, out_force=True)
```

### Display the cropped geotiffs

Note that the loaded geotiffs appear pretty dark. This is because of the limited rendering functionality of `rasterio`. The pixel values of the both the datacubes and the geotiffs are in units of reflectance (range from 0.0 to 1.0). 

```python
import matplotlib.pyplot as plt
import rasterio as rast
from rasterio.plot import show

base_dir = r'G:\BBE\AGROBOT\Shared Work\Data\PikaImagery4_Reflectance\hs_process_demo\spatial_crop_many_gdf'
tif_list = hsbatch._recurs_dir(base_dir, search_ext='.tif', level=0)

gdf_filter = gdf.copy()
# gdf_filter = hsbatch.my_spatial_mod._overlay_gdf(hsbatch.io.spyfile, gdf)

fig, ax = plt.subplots(figsize=(15, 15))
gdf_filter.plot(ax=ax, alpha=0.9, column='Crop', facecolor='none', edgecolor='blue', linewidth=3)

left = 1e9
bottom = 1e9
right = -1e9
top = -1e9
for fname_tif in tif_list:
    tif = rast.open(fname_tif)
    show(tif.read(), ax=ax, transform=tif.transform, adjust='linear')
    if tif.bounds[0] < left:
        left = tif.bounds[0]
    if tif.bounds[1] < bottom:
        bottom = tif.bounds[1]
    if tif.bounds[2] > right:
        right = tif.bounds[2]
    if tif.bounds[3] > top:
        top = tif.bounds[3]

# bbox = gdf_filter.total_bounds
plt.xlim((left-5, right+5))
plt.ylim((bottom-10, top+10))
```

## 2. Spectrally clip noisy bands

The bands that have relatively low signal can be clipped/removed to minimize the amount of noise in the spectral domain of the data.

### Plot a pixel before spectral clipping

We can open a datacube that was just spatially cropped and plot a single pixel across the spectral domain to observe the "noisy" spots: 

```python
import numpy as np
import os
import seaborn as sns

from hs_process import hsio

fname_clip = os.path.join(base_dir, 'Wells_rep2_20180711_18h19m_pika_gige_7_613-spatial-crop.bip')
io = hsio(fname_clip)  # hsio is a module that aids in reading/writing (predominatly uses the Spectral Python package)
array = io.spyfile.load()  # load the data from the Spectral Python (SpyFile) object
meta_bands = io.tools.meta_bands  # Get a dictionary of the band number (keys) and wavelengths (values)
pix = array[15, 40, :]  # Choose the data for a single pixel (3rd dimension of the array)
x = pd.Series(meta_bands, name='Wavelength (nm)')  # add wavelength info to Series for plotting on x-axis 
y = pd.Series(np.reshape(pix, (array.shape[2])) * 100, name='Reflectance (%)')  # reshape data so its a single dimension and add to Series for plotting
sns.lineplot(x, y)  # generate a line plot of the pixel across the spectral domain
```

Spectrally clipping is a simple process using `hs_process.batch.spectral_clip()`; the only parameter that must be defined is `wl_bands`, which indicates the wavelengths that are to be clipped from the dataset, and should be the minimum and maximum wavelenths to clip from the datacube. If multiple groups of wavelengths should be clipped, `wl_bands` can be a list of lists.

### Examples:

 - `wl_bands = [760, 776]` will clip all bands greater than 760.0 nm and less than 776.0 nm
 - `wl_bands = [[0, 420], [760, 776], [813, 827], [880, 1000]]` will clip all bands less than 420.0 nm, bands greater than 760.0 nm and less than 776.0 nm, bands greater than 813.0 nm and less than 827.0 nm, and bands greater than 880 nm (this is the default setting).

```python
base_dir = r'G:\BBE\AGROBOT\Shared Work\Data\PikaImagery4_Reflectance\hs_process_demo\spatial_crop_many_gdf'

hsbatch = batch(base_dir, search_ext='.bip')
fname_list = hsbatch.fname_list
folder_name='spec_clip'
hsbatch.spectral_clip(folder_name=folder_name, name_append='spec-clip',
                      wl_bands=[[0, 420], [760, 776], [813, 827], [880, 1000]],
                      out_force=True)
```

### Plot a pixel after spectral clipping:

```python
fname_clip = os.path.join(base_dir, folder_name, 'Wells_rep2_20180711_18h19m_pika_gige_7_613-spec-clip.bip')
io = hsio(fname_clip)
array = io.spyfile.load()
meta_bands = io.tools.meta_bands
pix = array[15, 40, :]
x = pd.Series(meta_bands, name='Wavelength (nm)')
y = pd.Series(np.reshape(pix, (array.shape[2])) * 100, name='Reflectance (%)')
sns.lineplot(x, y)
```

## 3. Spectrally smooth each pixel

Even though the noisiest bands have been removed, there is still some degree of noise present. This is apparent when looking at the "choppiness" of the spectra in the previous line plot. Spectrally smoothing is achieved by applying the [Savitzky-Golay filter](https://en.wikipedia.org/wiki/Savitzky%E2%80%93Golay_filter) using `hs_process.batch.spectral_smooth()`. Besides I/O information, the only parameters to set are the `window_size` (size of the moving window across the spectral domain) and `order` (polynomial order to fit the smoothed dataset). We will use an 11 band window size and 2nd order polynomial:

```python
base_dir = os.path.join(base_dir, folder_name)
folder_name = 'spec_smooth'
df_smooth_stats = hsbatch.spectral_smooth(
        base_dir=base_dir, folder_name=folder_name,
        name_append='spec-smooth', window_size=11, order=2, stats=True,
        out_force=True)
```

### Plot a pixel after spectral smoothing:

```python
fname_smooth = os.path.join(base_dir, folder_name, 'Wells_rep2_20180711_18h19m_pika_gige_7_613-spec-smooth.bip')
io = hsio(fname_smooth)
array = io.spyfile.load()
meta_bands = io.tools.meta_bands
pix = array[15, 40, :]
x = pd.Series(meta_bands, name='Wavelength (nm)')
y = pd.Series(np.reshape(pix, (array.shape[2])) * 100, name='Reflectance (%)')
sns.lineplot(x, y)
```

## 4. Segmenting the vegetation

Now that the noise is removed from the spectral domain, we can focus on masking out the pixels that we are not interested in (e.g., soil and shadow). This can be achieved a number of ways, but we will do it in two separate steps:
1. Perform a k-means clustering algorithm on each image across the entire spectral domain (2-3 classes), then mask out the class with the lowest NDVI value (NDVI is indicative of vegetative fraction). The goal here is to mask out areas of the image with relatively low reflectance across the spectral domain (soil and shadow) - shadows can be particularly challenging because they will tend to pass through a filter that attempts to striclty separate vegetation from soil (and that's not good). 
2. Apply the [Normalized Difference Red-Edge Index](https://en.wikipedia.org/wiki/Normalized_Difference_Red_Edge_Index) and maks out the lowest 90% of pixels in the image.

The masks that result from each of these steps will be combined (i.e., if either mask has been set for a given pixel, that pixel will be excluded from the dataset.

### Segmentation 1: K-means
`hs_process.batch.segment_kmeans()` will perform k-means clustering on each datacube in the directory. This step may be a bit time consuming if using more than just a few classes. The relevant parameters that can be passed are:
 - `n_classes`: the number of classes to cluster.
 - `max_iter`: the maximum number of iterations before terminating the clustering algorithm if convergence has not been met.
 - `plot_out`: whether to save a line plot of the spectra for each class.

```python
base_dir = os.path.join(base_dir, folder_name)
folder_name = 'kmeans'

hsbatch.segment_kmeans(base_dir=base_dir, folder_name=folder_name,
                       name_append='kmeans', geotiff=True,
                       n_classes=5, max_iter=100, plot_out=True,
                       out_force=True)
```

### Segmentation 2: Compute MCARI2 spectral index
We want to mask out the lowest 90% of pixels in the image after applying the Modified Chlorophyll Absorption Index 2 (MCARI2), so we will use `hs_process.batch.segment_band_math()` to compute the MCARI2 for each datacube. The relevant parameters that can be passed are:
 - `method`: must be either "mcari2" (modified chlorophyll absorption index), "ndi" (normalized difference index), "ratio" (simple ratio index), or "derivative" (derivative-type index); indicates the type of spectral index to apply.
 - `wl1`: the wavelength (or set of wavelengths) to be used as the first parameter of the band math index; if passed as a `list`, all bands in the list are consolidated by calculating the mean pixel value across those bands.
 - `wl2`: the wavelength (or set of wavelengths) to be used as the second parameter of the band math index; also can be passed as a list in the same way as `wl1`.
 - `wl3`: the wavelength (or set of wavelengths) to be used as the second parameter of the band math index; also can be passed as a list in the same way as `wl1`.

```python
# using same directory and going to save alongside k-means
folder_name = 'band_math'
method = 'mcari2'
wl1 = 800
wl2 = 670
wl3 = 550

hsbatch.segment_band_math(base_dir=base_dir, folder_name=folder_name,
                          name_append='band-math', geotiff=True, method=method,
                          wl1=wl1, wl2=wl2, wl3=wl3, plot_out=True,
                          out_force=True)
```

Notice that we passed `plot_out=True` in both the `segment_kmeans()` function and `segment_band_math()` function. The spectra of each K-means class for every image was plotted and saved as a .png image:

A histogram was also created for the band math spectral index (MCARI2) and saved as a .png image:

### Segmentation 3: Combining K-means results and MCARI2 results to create a composite mask

When the K-means clustering concluded, a '.csv' file was saved to the output directory (in the "kmeans" folder). This file contains the results of the K-means clustering algorithm for every image, including the number of pixels that fell in each cluster, as well as the mean NDIV, GNDVI, NDRE, and MCARI2 values in each cluster. This may be informative for data exploration and deciding how to proceed to apply the mask to the imagery for segmentaion. Let's load the file and have a look:

```python
fname_sheet = r'G:\BBE\AGROBOT\Shared Work\Data\PikaImagery4_Reflectance\hs_process_demo\spatial_crop_many_gdf\spec_clip\spec_smooth\kmeans\kmeans-stats.csv'
pd.read_csv(fname_sheet)
```

The `hsbatch.combine_kmeans_bandmath()` function achieves four main goals. For each image, it:
1. Finds the K-means cluster with the lowest spectral values and masks out all pixels that fall within that cluster. The spectral values it uses as a basis is set using the `kmeans_filter` parameter (can be one of "ndvi", "gndvi", "ndre", or "mcari2"). If `kmeans_mask_classes` is set to a number greater than one (which is the default), that number of clusters will be masked out (in order from lowest to highest spectral value.
2. Loads the relevant band math images (e.g., MCARI2 images in the "band_math" folder), and creates a second mask based on the `mask_percentile`, `mask_thresh`, and `mask_side` parameters (see documentation for more details).
3. These two masks are consolidated into a single mask. If a given pixel was masked in either of the first two steps, it is excluded from here on out.
4. The consolidated mask is applied to the datacube and is saved; the mean spectra of the unmasked pixels is also saved.

The most relevant parameters that can be passed are:
- `kmeans_mask_classes`: the number of K-means classes to mask from the datacube. By default, the classes with the lowest average spectral value (e.g., NDVI, GNDVI, MCARI2, etc.; based on `kmeans_filter` parameter) will be masked (default: 1).
- `kmeans_filter`: the spectral index to base the K-means mask on. Must be one of 'ndvi', 'gndvi', 'ndre', or 'mcari2'. Note that the K-means aglorithm does not use the in its clustering algorithm (default: 'mcari2').
- `mask_percentile`: the percentile of remaining pixels to mask based on the band math image (range: 0-100).
- `mask_side`: the side of the threshold or percentile for which to apply the mask, and must be either 'lower' or 'upper'; if 'lower', everything below the threshold/percentile will be masked (e.g., if `percentile`=90 and `side`='lower', the lowest 90% of pixels will be masked following the band math operation indicated by the `method`, `wl1`, and `wl2` parameters.

```python
folder_name = 'mask_combine'

kmeans_mask_classes = 2
kmeans_filter = 'mcari2'
mask_percentile = 90
mask_side='lower'

hsbatch.combine_kmeans_bandmath(fname_sheet, base_dir_out=None,
                                folder_name=folder_name,
                                name_append='mask-combine',
                                geotiff=True,
                                kmeans_mask_classes=kmeans_mask_classes,
                                kmeans_filter=kmeans_filter,
                                mask_percentile=mask_percentile,
                                mask_side=mask_side, out_force=True)
```

### Plot the masked mean vegetation spectra
That concludes the processing pipeline. At this point, we are left with a segmented image that should represent predominantly pure vegetation pixels, as well as the mean spectra for those segmented pixels. This data should be apporpriate for subsequent hyperspectral vegetation analysis.

```python
fname_spec = os.path.join(base_dir, folder_name, 'Wells_rep2_20180711_18h19m_pika_gige_7_613-mask-combine-spec-mean.spec.hdr')
hsbatch.io.read_spec(fname_spec)
array = hsbatch.io.spyfile_spec.load()  # load the data from the Spectral Python (SpyFile) object
meta_bands = hsbatch.io.tools.meta_bands  # Get a dictionary of the band number (keys) and wavelengths (values)
x = pd.Series(meta_bands, name='Wavelength (nm)')  # add wavelength info to Series for plotting on x-axis 
y = pd.Series(np.reshape(array, (array.shape[2])) * 100, name='Reflectance (%)')  # reshape data so its a single dimension and add to Series for plotting
sns.lineplot(x, y).set_title(hsbatch.io.name_short)  # generate a line plot of the pixel across the spectral domain
```
